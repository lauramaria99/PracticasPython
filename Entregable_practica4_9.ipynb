```
import matplotlib.pyplot as plt
import numpy as np
```

**Representación de señales**

```
Fo = 40 # Frecuencia fundamental de la señal
Tp = 1/Fo # Periodo de la señal
Fs = 1000 # Frecuencia de muestreo
T = 1/Fs # Periodo de muestreo
t = np.arange(0, Tp+T, T) # Tiempo para un ciclo de la señal con duración de Tp más una muestra T
A = 5
x = A*np.sin(2*np.pi*Fo*t)
plt.plot(t, x, marker='o')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

```
fo = Fo/Fs
n = np.arange(0, len(t))
x1 = A*np.sin(2*np.pi*fo*n)
plt.plot(n, x1, marker='o')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.show()
```

a) Compruebe que los valores de las primeras cinco muestras son los mismos.

```

print('Muestras de la señal en el dominio del tiempo: ', x[0:5])
print('Muestras de la señal en el dominio de la frecuencia: ',x1[0:5])
```

Se observa que al imprimir los cinco primeros valores de las señales en el dominio del tiempo y la frecuencia estos son iguales. 

**Energía y potencia en el dominio del tiempo**

```
energia = sum(x**2)
potencia = energia/(len(t)-1)
rms = np.sqrt(potencia)
t10 = np.arange(0,10*Tp+T,T)
x10 = A*np.sin(2*np.pi*Fo*t10)
energia10 = sum(x10**2)
potencia10 = energia10/(len(t10)-1)
rms10 = np.sqrt(potencia)
print('La energía de un ciclo es: ',energia)
print('La energía de 10 ciclos es: ',energia10)
print('La potencia de un ciclo es: ',potencia)
print('La potencia de 10 ciclos es: ',potencia10)
print('El valor rms de un ciclo es: ',rms)
print('El valor rms de 10 ciclos es: ',rms10)
```

b) La potencia es la misma que en x(n). ¿Por qué?

La  potencia es la misma, esto se debe a que como la señal en cuestión es periodica, la potencia media de esta se puede hallar integrando un solo periodo de la señal, luego será lo mismo hacerlo sobre la señal completa o solo para un ciclo de esta.

**Análisis de Fourier en tiempo discreto**

```
X10 = np.fft.fft(x10)
plt.plot(abs(X10))
plt.title('Módulos de la DFT')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.show()
```

```
N = len(X10)
F = np.arange(0,N)*Fs/N
plt.plot(F,abs(X10))
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.show()
```

```
Nmitad = int(np.ceil(N/2))
Fmitad = np.arange(0,Nmitad)*Fs/N
X10mitad = X10[0:Nmitad]
plt.plot(Fmitad,abs(X10mitad))
plt.ylabel('Amplitud')
plt.xlabel('Frecuencia (Hz)')
plt.show()
```

```
plt.plot(Fmitad,abs(X10mitad))
plt.ylabel('Amplitud')
plt.xlabel('Frecuencia (Hz)')
plt.xlim(0,80)
plt.show()
```
